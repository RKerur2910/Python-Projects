```python
player_1 = "Sam"
player_2 = "Alex"
current_round = 1

print("Game On!")
print(f"player 1: {player_1}")
print(f"player 2: {player_2}")
print("--------------------------")

#Round 1
print(f"Round {current_round}!")
player_1_score = 18
player_2_score = 13
print(f"{player_1} wins with {player_1_score} to {player_2_score}")
print("--------------------------")

#Round 2
print(f"Round {current_round + 1}!")
player_1_score = 25
player_2_score = 30
print(f"{player_2} defeats {player_1} by {player_2_score - player_1_score} points")
```
